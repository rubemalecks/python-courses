# Classes

```python
class Pessoa:
    pass

p1 = Pessoa()
p2 = Pessoa()

# p1 != p2

p1.nome = "Luiz"

print(p1.nome) 
```

```python
print(p2.nome) # Erro de atributo
```

Variáveis de instancia (ou variáveis de classe) são chamadas de atributos da classe.
No caso ..."nome" é um atributo da classe p1 

Para que p2 tenha esse atributo ...

```python
p2.nome = 'João'
```

```python
print(p1.nome)
print(p2.nome)
```

A classe deve ser dinâmica se modelando a situação que está sendo usada, sendo assim podemos fazer isto de forma mais simples ... continuando ...
